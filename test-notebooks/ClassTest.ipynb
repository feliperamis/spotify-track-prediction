```python
import pandas as pd
import numpy as np
import sklearn
import sklearn.model_selection as skl
import matplotlib.pyplot as plt
import json

df = pd.read_csv('../datasets/SpotifyAudioFeatures-full.csv')

data_X = df.drop(['popularity'], axis = 1)
data_Y = df['popularity']

df.head()
```

```python
Xn = pd.get_dummies(data_X) # All numerical to categorical
Xn.head()
```

```python
# Separate training to test. We shuffle to desorder the data
from sklearn.utils import shuffle
Xn, Y = shuffle(Xn, data_Y)

(x_train, x_test, y_train, y_test) = skl.train_test_split(Xn, Y, test_size=0.33)
```

```python
from sklearn.naive_bayes import BernoulliNB # For binary features (every variables)
from statsmodels.stats.proportion

model = BernoulliNB()
# confusion_matrix
# classification_report
```

```python
scores = sklearn.model_selection.cross_val_score(model, Xn, Y, cv=10)
```

```python
# Decision tree
from sklearn import tree 

clf = tree.DecisionTreeClassifier(criterion='entropy')
pred = clf.fit(x_train, y_train).predict(x_test)
print(f'{sklearn.metrics.confusion_matrix(y_test, pred)}\n')
print(f'Accuracy on test set: {sklearn.metrics.accuracy_score(y_test, pred)}\n')
print(sklearn.metrics.classification_report(y_test, pred))
```
